```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Provided details for the adult dataset
adult_mean_height = 150
adult_std_dev_height = 10
adult_mean_weight = 50
adult_std_dev_weight = 5


# Provided details for the child dataset
child_mean_height = 50
child_std_dev_height = 10
child_mean_weight = 25
child_std_dev_weight = 5

# Create target arrays
target_1 = np.zeros(1000)  # Adults
target_2 = np.ones(1000)   # Children
```

```python
# Generate adult data points with the provided distribution details
adult_heights = np.random.normal(loc=adult_mean_height, scale=adult_std_dev_height, size=1000)
adult_weights = np.random.normal(loc=adult_mean_weight, scale=adult_std_dev_weight, size=1000)
adult_data = np.column_stack((adult_heights, adult_weights))

# Generate child data points with the provided distribution details
child_heights = np.random.normal(loc=child_mean_height, scale=child_std_dev_height, size=1000)
child_weights = np.random.normal(loc=child_mean_weight, scale=child_std_dev_weight, size=1000)
child_data = np.column_stack((child_heights, child_weights))
```

```python
# Merge both datasets and targets
data = np.vstack((adult_data, child_data))
targets = np.concatenate((target_1, target_2))

print(data.shape)
```

```python
# Convert to Pandas DataFrame and save as CSV files
df_data = pd.DataFrame(data, columns=['Height', 'Weight'])
df_targets = pd.DataFrame(targets, columns=['Target'])

# Define file paths
data_csv_path = 'dataset.csv'  # This will save in the current working directory
targets_csv_path = 'targets.csv'

# Save DataFrames to CSV files
df_data.to_csv(data_csv_path, index=False)
df_targets.to_csv(targets_csv_path, index=False)
```

```python
# Visualize the data in a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(adult_data[:, 0], adult_data[:, 1], c='blue', label='Adults', alpha=0.5)
plt.scatter(child_data[:, 0], child_data[:, 1], c='red', label='Children', alpha=0.5)
plt.title("Simulated Height and Weight Data")
plt.xlabel("Height (in cm)")
plt.ylabel("Weight (in kg)")
plt.legend()
plt.grid(True)
plt.show()
```
