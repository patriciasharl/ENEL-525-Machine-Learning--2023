```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#initialize input, weight vector and bias
p=np.array([[1,2,3,1,2,4],[4,5,3.5,0.5,2,0.5]])
W=np.array([[0,0]])
b=0
a=np.array([0,0,0,0,0,0])
t=np.array([1,1,1,0,0,1])
e=np.array([0,0,0,0,0,0])
flag=np.array([1,1,1,1,1,1])

```

```python
#Hardlim function

def hardlim(n):
    if n>=0:
        return 1
    else:
        return 0
    
#error function
def error(n):
    if e[n] != 0:
        flag[n]=1
    else:
        flag[n]=0
    
```

```python
while flag.any()!=0:
    #flag=np.array([1,1,1,1,1,1])
    for i in range (6):
        n=np.dot(W,p[:,i])
        a[i]=hardlim(n+b)
        e[i]= t[i]-a[i]
        print ("e[",i,"] is equal to",e[i])
        if e[i]==0:
            flag[i]=0
            i=i+1
        else:
            flag[i]=1
            W=W+(e[i]*p[:,i].T)
            print ("W is equal to",W)
            b= b+e[i]
            print ("b is equal to",b)
            break
        
    print ("flag is equal to",flag)
print ("b is equal to",b) 
            
    
    
```

```python

# Plot grid
plt.grid(color='g', linestyle='-', linewidth=0.1)
plt.xlim(-1, 6)

# Plot class 1 samples
plt.plot([1, 2, 3,4], [4, 5, 3.5,0.5], 'x', label='Class 1')

# Plot class 2 samples
plt.plot([1, 2], [0.5, 2], 'o', label='Class 0')

# Plot the boundary line
x = [-1, 6]  
#slope intercept form:
#y = (-(b / w2) / (b / w1))x + (-b / w2)
m =(-(b/W[0,1])/(b/W[0,0]))
b = (-b/W[0,1])
y = [m * xi + b for xi in x]

# Plot the boundary line
plt.plot(x, y, label='Decision Boundary', color='red', linestyle='--',linewidth=1)

# Set axis labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```
